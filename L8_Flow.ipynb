```python
from crewai.flow.flow import Flow, listen, start
from IPython.display import display

def log(msg: str) -> None:
    """Show message in Jupyter output (works inside CrewAI flow steps)."""
    display(msg)
```

```python
class UnstructuredStateFlow(Flow):

    def __init__(self):
        super().__init__()
        self.data = "Hello, world!"
        self.data2 = "First flow example"
        
    @start
    def initialize_data(self):
        log("Initializing data")
        self.data = "Hello, world!"
        self.data2 = "First flow example"
        return "initializing data"
        

    @listen(initialize_data)
    def processed_data(self):
        log(f"Unstructured state: {self.data}")
        log(f"Unstructured state: {self.data2}")
        self.data = "Hello, world! 2"
        return "processed data"
    
    def show_data(self):
        log(self.data)
        log(self.data2)

```

```python
flow = UnstructuredStateFlow()
# In Jupyter, use kickoff_async + await (kickoff() uses asyncio.run() which fails when a loop is already running)
result = await flow.kickoff_async()
# log(result)

log(flow.show_data())
```
