```python
llm = "ollama/llama3.2"
```

```python
from pathlib import Path

def load_knowledgebase(filename: str) -> str:

        knowledge_dir = Path("./knowledge")
        knowledge_dir.mkdir(exist_ok=True)

        file_path = knowledge_dir / filename
        text = file_path.read_text(encoding="utf-8")
        max_chars = 6_000
        text = text[:max_chars] + ("\n[...]" if len(text) > max_chars else "")

        return text
   

```

```python
# Create agent with read_knowledge tool (RAG) â€” knowledge is already saved in the scrape cell above
from crewai import Agent, Task, Crew
researcher = Agent(
    role='latest trends Analyst',
    goal='Provide the top 3 AI trends by using RAG: reference the already-saved knowledge base (knowledge/deeplearning_batch.txt from The Batch) via your read tool; do not save again.',
    backstory='You use retrieval-augmented generation (RAG): the knowledge was saved beforehand in the notebook. Use your read_knowledge tool to reference that content and produce a numbered list of 3 latest AI trends.',
    llm=llm,
    verbose=True
)
```

```python
text = load_knowledgebase('deeplearning_batch.txt')  # this is the RAG content
rag_content = text
research = Task(
    description=(
        'Below is the RAG content (from text = load_website(url)). Use it to pick the 3 most notable recent AI trends and output ONLY a numbered list.\n\n'
        '--- RAG CONTENT (text = load_website(url)) ---\n' + rag_content + '\n--- END ---\n\n'
        'Output format: exactly 3 lines, each starting with 1. or 2. or 3. followed by trend name and 1â€“2 sentences. No other text.'
    ),
    expected_output='A numbered list of exactly 3 items: 1. [Trend]: [1-2 sentences]  2. [Trend]: [1-2 sentences]  3. [Trend]: [1-2 sentences]',
    agent=researcher,
)
```

```python
# Assemble a crew with planning enabled

crew = Crew(
    agents=[researcher],
    tasks=[research],
    verbose=True,
    planning=False,  # Enable planning feature
    llm="ollama/llama3.2",   # for planner
)

# Execute tasks
result = crew.kickoff()
```
